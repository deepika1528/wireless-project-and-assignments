```python
import matplotlib.pyplot as plt

# Data from simulation
nodes = [75, 85, 100, 150, 200]

packets_sent = [136, 208, 211, 140, 265]
packets_received = [45, 112, 92, 68, 239]

delay = [0.251682, 0.0581441, 0.0732374, 0.0640389, 0.0648173]
jitter = [0.0973333, 0.00993213, 0.0526254, 0.016957, 0.0209904]

# Packets Graph
plt.plot(nodes, packets_sent, marker='o', label='Sent')
plt.plot(nodes, packets_received, marker='o', label='Received')
plt.xlabel('Number of Nodes')
plt.ylabel('Packets')
plt.title('Packets Sent vs Received')
plt.legend()
plt.grid()
plt.show()

# PDR Calculation + Graph
pdr = [(r/s)*100 for r, s in zip(packets_received, packets_sent)]

plt.plot(nodes, pdr, marker='o')
plt.xlabel('Nodes')
plt.ylabel('PDR (%)')
plt.title('Packet Delivery Ratio')
plt.grid()
plt.show()

# Delay Graph
plt.plot(nodes, delay, marker='o')
plt.xlabel('Nodes')
plt.ylabel('Delay (sec)')
plt.title('End-to-End Delay')
plt.grid()
plt.show()

# Jitter Graph
plt.plot(nodes, jitter, marker='o')
plt.xlabel('Nodes')
plt.ylabel('Jitter (sec)')
plt.title('Jitter Analysis')
plt.grid()
plt.show()
```

## Analysis

- Packet Delivery Ratio (PDR) varies with node density due to changing contention levels.
- Delay fluctuates based on traffic load and retransmissions.
- Jitter shows variation due to inconsistent transmission timing.
- The proposed adaptive cross-layer model maintains better stability under moderate network conditions.
